To simulate a p-value with a 50/50 coin flipping model, we are testing the null hypothesis that people don't have a preference for tilting their heads to the right or left when kissing. According to the null hypothesis, the probability of tilting right or left should be 50/50, just like flipping a fair coin.

Steps to simulate the p-value:
1. Observations:
    - It is observed that 80 out of 124 couples tilt their heads to the right when kissing, 64.5%. 
    - We then calculate how unusual this proportion is if the true probability of tilting right is 50% based on the null hypothesis.
2. Simulating null hypothesis
    - Under the null hypothesis, each couple has a 50% chance of tilting right
    - We simulate this scenario by flipping a virtual coin 124 times, representing each couple's decision to tilt left or right. Each trial is independent, and the probability of a right tilt is 50%.
    - Repeat this simulation many times to generate the distribution of possible outcomes.
3. Count extreme results: 
    - After simulating the head tilts for 124 couples multiple times, we count how often we get 80 or more right tilts
4. calculate the p-value
    - The p-value is the proportion of simulations where 80 or more couple tilt their heads to the right.
    - let's say we simulate 10000 trials where 124 coins are flipped each time. Let's assume that in 500 of those trails, 80 or more heads occur
    - p-value = (number of simulations with 80 or more right tilts)/(total number of simulations)
        p-value = 500/10,000 = 0.05


Our simulated p-value is 0.05, which according to the chart, there is moderate evidence against the null hypothesis. This means there is a reasonable chance that the observed 64.5% head tilting to the right is not just due to random chance, and there may indeed be a right tilt preference when kissing.

```python
import numpy as np
import plotly.figure_factory as ff
import plotly.graph_objects as go

# Simulated bootstrapped statistics under H0 (two-sided test)
IncreaseProportionSimulations_underH0random = np.random.normal(loc=0, scale=0.1, size=1000)
group_labels = ['Bootstrap<br>Sampling<br>Distribution<br>of the<br>Sample<br>Mean<br><br>assuming<br>that the<br>H0 null<br>hypothesis<br>IS TRUE']

# Observed test statistic (for demonstration purposes)
observed_test_statistic = (np.random.normal(0.05, 0.1, 1000) > 0).mean()

# Population parameter value under H0
population_parameter_value_under_H0 = 0  # Assuming the null hypothesis is zero

# Number of simulations
number_of_simulations = len(IncreaseProportionSimulations_underH0random)

# Calculate the p-value (Two-tailed test)
SimTestStats_as_or_more_extreme_than_ObsTestStat = \
    abs(IncreaseProportionSimulations_underH0random - population_parameter_value_under_H0) >= \
    abs(observed_test_statistic - population_parameter_value_under_H0)

p_value = (SimTestStats_as_or_more_extreme_than_ObsTestStat).sum() / number_of_simulations

# Print the result
print("Number of Simulations: ", number_of_simulations, "\n\n",
      "Number of simulated test statistics (under H0)\n",
      'that are "as or more extreme" than the observed test statistic: ',
      SimTestStats_as_or_more_extreme_than_ObsTestStat.sum(), "\n\n",
      'p-value\n(= simulations "as or more extreme" / total simulations): ', p_value, sep="")

# Create distribution plot
hist_data = [IncreaseProportionSimulations_underH0random + np.random.uniform(-0.05, 0.05, size=len(IncreaseProportionSimulations_underH0random))]
fig = ff.create_distplot(hist_data, group_labels, curve_type='normal',
                         show_hist=True, show_rug=False, bin_size=0.1)

# Add observed test statistic line
pv_y = 2.5
pv_y_ = .25
fig.add_shape(type="line", x0=observed_test_statistic, y0=0, 
              x1=observed_test_statistic, y1=pv_y,
              line=dict(color="Green", width=4), name="Observed Test Statistic")
fig.add_trace(go.Scatter(x=[observed_test_statistic], y=[pv_y+pv_y_], 
                         text=["Observed<br>Test Statistic<br>^"], mode="text", showlegend=False))

# Add symmetric observed test statistic for two-tailed test
symmetric_test_statistic = population_parameter_value_under_H0 -\
                           abs(observed_test_statistic - population_parameter_value_under_H0)
fig.add_shape(type="line", x0=symmetric_test_statistic, y0=0, 
              x1=symmetric_test_statistic, y1=pv_y,
              line=dict(color="Green", width=4), name="Symmetric Test Statistic")
fig.add_trace(go.Scatter(x=[symmetric_test_statistic], y=[pv_y+pv_y_], 
                         text=['"Symmetric" Observed Test Statistic<br>^'], mode="text", showlegend=False))

# Add transparent rectangles for both extreme regions (two-tailed test)
fig.add_shape(type="rect", x0=-0.25, y0=0, x1=symmetric_test_statistic, y1=pv_y,
              fillcolor="LightCoral", opacity=0.5, line_width=0)
fig.add_shape(type="rect", x0=observed_test_statistic, y0=0, x1=1.25, y1=pv_y,
              fillcolor="LightCoral", opacity=0.5, line_width=0)

# Update layout
fig.update_layout(
    title="Bootstrapped Sampling Distribution<br>under H0 with p-value regions",
    xaxis_title="Mean Health Score Change", yaxis_title="Density", yaxis=dict(range=[0, pv_y+2*pv_y_]))

# Show the figure
fig.show(renderer="png")  # Use for GitHub/MarkUs submissions
```

3. Should We Expect the P-value to Be Smaller in a One-Tailed Test?
- Yes, we should expect the p-value to be smaller in a one-tailed test compared to a two-tailed test, assuming the observed test statistic is in the direction of interest.
    - In a two-tailed test, you are looking at extreme values in both directions (greater than and less than), so the p-value is effectively the probability of obtaining an extreme value in either direction.
    - In a one-tailed test, you're only interested in the extreme values in one direction (either greater or less), so the p-value is based on just one tail of the distribution.
    

Because the p-value in a one-tailed test is based on only one tail, it is typically smaller . This is why one-tailed tests are often more powerful when you have a clear hypothesis about the direction of the effect.

This means the probability of correctly identifying which was poured first is 50%.(The students are just guessing, and there’s no actual skill involved in choosing whether the tea or milk was poured first)
    - Alternative Hypothesis (HA): The students are not guessing and have some ability to distinguish whether the milk or tea was poured first. This suggests the probability of correctly identifying is greater than 50%.
- Quantitative Analysis
    - To test the validity of the null hypothesis, you can conduct a one-sample hypothesis test for proportions:
        1. Set the significance level (α), commonly 0.05.
        2. Calculate the sample proportion:
        - sample proportion = 49 / 80 = 0.6125
        3. Define the null hypothesis proportion (p0):
        - p0 = 0.5
        4. Compute the test statistic (z-score):
        - The test statistic for a one-sample proportion z-test is given by:
        - z = (sample proportion  - null hypothesis proportion) / root((null hypothesis proportion)(1 - null hypothesis proportion) / sample size)
        5. Calculate the p-value: Based on the z-score, find the p-value. If the p-value is less than α, you reject the null hypothesis.

```python
import numpy as np
from scipy.stats import norm

# Given data
n = 80
x = 49
p_hat = x / n
p_0 = 0.5
alpha = 0.05

# Standard error
se = np.sqrt(p_0 * (1 - p_0) / n)

# Test statistic
z = (p_hat - p_0) / se

# p-value for one-sided test (upper tail)
p_value = 1 - norm.cdf(z)

# Output results
print(f"Test statistic (z): {z}")
print(f"p-value: {p_value}")

# Conclusion
if p_value < alpha:
    print("Reject the null hypothesis. There is evidence that students have an ability to identify the correct order.")
else:
    print("Fail to reject the null hypothesis. There is no evidence that students can identify the correct order.")

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
n = 80
x = 49
p_hat = x / n
p_0 = 0.5
alpha = 0.05

# Standard error
se = np.sqrt(p_0 * (1 - p_0) / n)

# Test statistic
z = (p_hat - p_0) / se

# Create a range of z-values
z_values = np.linspace(-3, 3, 1000)

# Calculate the normal distribution for those z-values
normal_distribution = norm.pdf(z_values)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(z_values, normal_distribution, label='Standard Normal Distribution', color='blue')

# Fill the area for the p-value
x_fill = np.linspace(z, 3, 100)
plt.fill_between(x_fill, norm.pdf(x_fill), 0, alpha=0.5, color='red', label='P-value Area')

# Mark the test statistic
plt.axvline(z, color='green', linestyle='--', label=f'Test Statistic (z = {z:.2f})')

# Add labels and title
plt.title('Normal Distribution with Test Statistic')
plt.xlabel('Z-value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.xlim(-3, 3)
plt.ylim(0, max(normal_distribution) * 1.1)

# Show the plot
plt.show()
```

### Ethical Professionalism Considerations
    
Using p-values and hypothesis testing appropriately is an important ethical and professional responsibility of anyone doing data analysis. Actually, there is quite the quiet Contra-Versy (or is it Con-TROV-ersy?) around p-values. First, on a general level, it seems quite clear that p-values and hypothesis testing methodologies MUST play some ongoing contributing role in the so-called "replication crisis" rampantly afflicting mordern science; namely, "significant findings" made in scientific studies are not able to be reproduced by future studies at an alarming rate; and, this whole paradigm of "significant findings" is based on p-values and hypothesis testing... so, something's going on with this methodology in some way...
    
More specifically however, p-values are themselves quite problematic. To see this, just briefly consider the following article titles...

- [Why are p-values controversial?](https://www.tandfonline.com/doi/full/10.1080/00031305.2016.1277161) 
- [What a nerdy debate about p-values shows about science and how to fix it](https://www.vox.com/science-and-health/2017/7/31/16021654/p-values-statistical-significance-redefine-0005)
- [The reign of the p-value is over: what alternative analyses could we employ to fill the power vacuum?](https://royalsocietypublishing.org/doi/10.1098/rsbl.2019.0174)

- [Statistics experts urge scientists to rethink the p-value](https://www.spectrumnews.org/news/statistics-experts-urge-scientists-rethink-p-value)

While the issues here are relatively advanced and subtle (as introduced [here](https://www2.stat.duke.edu/~berger/p-values.html), presented [here](https://www.jarad.me/courses/stat587Eng/slides/Inference/I06-Pvalues/why_pvalues_dont_mean_what_you_think_they_mean.pdf), and demonstrated using simulation [here](https://jaradniemi.shinyapps.io/pvalue/)), the problem essentially comes down to the fact that most scientists (or just people) don't know how to really interpret the numeric value of a p-value. There are therefore two current proposed solutions to address this challenge.
    
1. Just interpreting p-values using the follwing table (which really isn't that hard, so it's surprising that this solution isn't more broadly adopted...)
    
|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|
    

2. Only do **hypothesis testing** on the basis of confidence intervals, not **p-values** (which might be the best solution wherever doing so is a realistic, convenient  possibility...)

    2. We do not prove $H_0$ is true, we instead do not have evidence to reject $H_0$: "We fail to reject the null hypothesis with a p-value of XYZ"<br><br>
